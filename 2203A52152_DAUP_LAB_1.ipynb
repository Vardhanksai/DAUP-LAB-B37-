```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
url = 'insurance.csv'
data = pd.read_csv(url)
```

```python
filtered_df = data[(data['age'] >= 20) & (data['age'] <= 40)]
plt.scatter(filtered_df['age'], filtered_df['smoker'], marker='+',c='green')
plt.title('Scatter Plot between Age (20 to 40) and Smoking')
plt.xlabel('Age')
plt.ylabel('Smoking')
plt.show()
```

```python
smokers_expenses = data[data['smoker'] == 'yes']['charges']
non_smokers_expenses = data[data['smoker'] == 'no']['charges']

plt.boxplot([smokers_expenses, non_smokers_expenses], labels=['Smokers', 'Non-Smokers'])
plt.title('Expenses of Smokers vs Non-Smokers')
plt.ylabel('Expenses')
plt.show()
```

```python
x_column = data.columns[0]
y_column = data.columns[2]

plt.figure(figsize=(10, 6))
plt.scatter(data[x_column], data[y_column], c='skyblue', marker='+', label='Data Points')
plt.title("Scatter Plot of {} vs {}".format(x_column, y_column))
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.legend()

for i in range(5):
    plt.text(data[x_column][i], data[y_column][i], str(i), fontsize=9)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numerical_cols = data.select_dtypes(include=['number']).columns.difference(['expenses'])

for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=col, y='charges', data=data)
    plt.title(f'Scatter Plot: {col} vs. Expenses')
    plt.xlabel(col)
    plt.ylabel('Expenses')
    plt.show()
```

```python
categorical_cols = data.select_dtypes(include=['object']).columns

for col in categorical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=col, y='charges', data=data)
    plt.title(f'{col.capitalize()} vs. Expenses')
    plt.xlabel(col.capitalize())
    plt.ylabel('Expenses')
    plt.show()
```

```python
import pandas as pd

gender_expenses = data.groupby('sex')['charges'].sum()

print(gender_expenses)

if gender_expenses['male'] > gender_expenses['female']:
    print("Males have more expenses.")
elif gender_expenses['female'] > gender_expenses['male']:
    print("Females have more expenses.")
else:
    print("Both genders have equal expenses.")
```

```python
smoker_gender_count = data[data['smoker'] == 'yes'].groupby('sex').size()

smoker_gender_count.plot(kind='bar',color = 'brown')
plt.title('Number of Smokers by Gender')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd

teenage_smokers = data[(data['age'] >= 13) & (data['age'] <= 19) & (data['smoker'] == 'yes')]
num_teenage_smokers = len(teenage_smokers)
print("Number of teenage smokers:", num_teenage_smokers)
```

```python
threshold = data[x_column].mean()
filtered_data = data[data[x_column] > threshold]

plt.figure(figsize=(10, 6))
plt.scatter(filtered_data[x_column], filtered_data[y_column], c='orange', marker='o', label='Filtered Data')
plt.title("Filtered Scatter Plot of {} vs {}".format(x_column, y_column))
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.legend()
```
